```python
#merge the two dataframe 
df3 = pd.DataFrame({ "city":["Delhi", "Mumbai", "Noida"], "Temp":[23,43,54]},)
df3
```

```python
df4 = pd.DataFrame({ "city":["Mumbai", "Delhi", "Noida"], "Humadity":[23,12,54]},)
df4
```

```python
df5 = pd.merge(df3, df4, on = "city")
df5
```

```python
df6 = pd.merge(df3, df4, on = "city", how = 'outer')
df6
```

```python
df7 = pd.merge(df3, df4, on = "city", how =  'right', indicator = True)
df7
```

```python
#now we will disccussed about the 
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
merge = pd.merge(df1,df2,on = 'city')
merge
```

```python
merge = pd.merge(df1,df2,on = 'city', how =  'outer')
merge
```

```python
merge = pd.merge(df1,df2,on = 'city', how =  'outer', indicator =  True)
merge
```

```python
merge = pd.merge(df1,df2,on = 'city', how =  'outer', suffixes = ("_first","_second"))
merge
```

```python
merge = pd.merge(df1,df2,on = 'city', how =  'outer', suffixes = ("_first","_second"))
merge.set_index('city', inplace = True)
merge
```
